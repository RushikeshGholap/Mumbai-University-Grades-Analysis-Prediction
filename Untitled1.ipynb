```python
counter = pd.read_csv("./csv_db/counter.csv")
mean_df = pd.read_csv('./csv_db/mean_df.csv')
max_df = pd.read_csv('./csv_db/max_df.csv')
min_df = pd.read_csv('./csv_db/min_df.csv')
template = pd.read_csv('./csv_db/template.csv')
fig = go.Figure()
sub_dep = 'COMPUTER ENGINEERING'
for sub_college in regular['college_code'].unique():
    
    
    mean_college = []
    max_college = []
    min_college = []
    for x in ['sem_1','sem_2','sem_3','sem_4','sem_5','sem_6','sem_7','sem_8','cgpi']:

            mean_college.append(float(mean_df[(mean_df['college_code']==sub_college) & (mean_df['department']==sub_dep)][x]))
            max_college.append(float(max_df[(max_df['college_code']==sub_college) & (max_df['department']==sub_dep)][x]))
            min_college.append(float(min_df[(min_df['college_code']==sub_college) & (min_df['department']==sub_dep)][x]))
    
    sem = ["SEM I","SEM II","SEM III","SEM IV","SEM V","SEM VI","SEM VII","SEM VIII","CGPI"]   
    # Add scatter trace for line
    fig.add_trace(go.Scatter(x=sem, y=max_college, name='Max of College '+sub_college,
                            line=dict( width=4)))
    fig.add_trace(go.Scatter(x=sem, y=mean_college, name='Average of College '+sub_college,
                            line=dict( width=4)))
    fig.add_trace(go.Scatter(x=sem, y=min_college, name = 'Min of College '+sub_college,
                            line=dict( width=4)))


    fig.update_layout(
        shapes=[
            # 1st highlight
            dict(
                type="rect",
                # x-reference is assigned to the x-values
                xref="x",
                # y-reference is assigned to the plot paper [0,1]
                yref="paper",
                x0="SEM I",
                y0=0,
                x1="SEM III",
                y1=1,
                fillcolor="LightGreen",
                opacity=0.5,
                layer="below",
                line_width=0,
            ),
            # 2nd highlight
            dict(
                type="rect",
                xref="x",
                yref="paper",
                x0="SEM III",
                y0=0,
                x1="SEM VI",
                y1=1,
                fillcolor="orangered",
                opacity=0.5,
                layer="below",
                line_width=0,
            ),
            # 3rd highlight 
            dict(
                type="rect",
                xref="x",
                yref="paper",
                x0="SEM VI",
                y0=0,
                x1="SEM VIII",
                y1=1,
                fillcolor="LightGreen",
                opacity=0.5,
                layer="below",
                line_width=0,
            )
        ]
    )

    fig.show()
```
